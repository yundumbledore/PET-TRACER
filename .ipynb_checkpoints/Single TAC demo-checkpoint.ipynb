```python
from PetTracer import *
import torch

device = torch.device('cuda:0' if torch.cuda.is_available() else 'mps' if torch.backends.mps.is_available() else 'cpu')
print(device)
```

```python
# Load model
model = get_model(device)
```

```python
# Process a TAC from the left frontal cortex
data_path = './Sample_data/brain_TAC.h5'
parameter_name = 'Ki'
title = 'Left Frontal Cortex'
Ki = demo_single_TAC(model, device, data_path)
plot_posterior(title, parameter_name, Ki)
```

```python
# Process a TAC from the myocardium
data_path = './Sample_data/myocardium_TAC.h5'
parameter_name = 'Ki'
title = 'Myocardium'
Ki = demo_single_TAC(model, device, data_path)
plot_posterior(title, parameter_name, Ki)
```
